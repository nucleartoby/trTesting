```python
TICKER = 'SPY'

df = yf.Ticker(TICKER).history(period="max")
df['MA'] = df['Close'].rolling(50).mean()

df = df.iloc[-100:, :] #last 100 row, all columns

plt.plot(df['Close']) 
plt.plot(df['MA'])
plt.title(f'{TICKER} Close vs. Moving Average')
```
